```python
import random
import matplotlib.pyplot as plt
#atividade de ME323
#porta dos desesperados
def main():
    n = 1000
    print("TESTANDO AS PROBABILIDADES COM 1000 JOGOS PARA CADA CASO")
    
    lista = bla(1, n)
    plt.subplot(211)
    plt.plot(lista, 'C4', label='Trocando de porta')
    plt.ylabel('Proporção acum. de vitórias')
    plt.xlabel('Número de jogos')
    plt.legend()
        
    lista = bla(0, n)
    plt.subplot(212)
    plt.plot(lista, 'C2', label='Não trocando de porta')
    plt.ylabel('Proporção acum. de vitórias')
    plt.xlabel('Número de jogos')
    plt.legend()

def bla(troca, n):
    lista = []
    teste = [0, 0, 0]
    ganhou = 0
    for j in range(1, n+1):
        premio = random.randint (0, 2)
        teste[premio] = 1
        
        escolha = random.randint (0, 2)
        
        for i in range(0, 2):
            if (i != escolha and teste[i] != 1):
                teste[i] = 'X'
                break

        if (troca == 1):
            for i in range(0, 2):
                if (i != escolha and teste[i] != 'X'):
                    escolha = i
                    break
        
        if (teste[escolha] == 1):
            ganhou += 1
            
        lista.append(ganhou/j)    
        teste = [0, 0, 0]
    return lista

main()


    
```
